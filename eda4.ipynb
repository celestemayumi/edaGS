```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Análise Exploratória

Análise que diz sobre o lixo plastico mal gerenciado per capita em kg por ano.

```python
# Carregando os dados

df = pd.read_csv('./datasets/4.csv')
```

Vendo informações sobre a tabela, analisando a necessidade de um possível tratamento e descobrir o que pode ser analisado.

### Início

```python
df_continent = df[df[' Código'].isna()].sort_values(by=' Lixo plástico mal gerenciado por pessoa (kg por ano)', ascending=False).head(10)
df_continent

#Nao tem a america do sul, acredito que nao seja bom o suficiente para gerar um grafico informativo devido a essa falta
```

```python
# Filtrando a tabela por países

df_countries = df[df[' Código'].notnull()].sort_values(by=' Lixo plástico mal gerenciado por pessoa (kg por ano)', ascending=False).head(10)
df_countries
```

## Lixo plástico mal gerenciado per capita ordenado por país (kg por ano)

```python
plt.figure(figsize=(15, 5))

sns.barplot(x=df_countries['Entidade'], y=df_countries[' Lixo plástico mal gerenciado por pessoa (kg por ano)'], palette= 'bright', hue= df_countries['Entidade'])

plt.title('Lixo plástico mal gerenciado por pessoa (kg por ano)')
plt.ylabel('')
plt.xlabel('')

plt.show()
```

```python
# Filtrando por países da América do Sul

df_sa = df[df['Entidade'].isin(['Brazil', 'Argentina','Paraguay','Uruguay','Peru','Chile','Colombia','Venezuela','Ecuador','French Guiana','Suriname','Trinidad and Tobago', 'Aruba','Guyana'])].sort_values(by=' Lixo plástico mal gerenciado por pessoa (kg por ano)', ascending=False)
df_sa
```

```python
plt.figure(figsize=(15, 5))

sns.barplot(x=df_sa['Entidade'], y=df_sa[' Lixo plástico mal gerenciado por pessoa (kg por ano)'], palette= 'bright', hue= df_sa['Entidade'])

plt.title('Lixo plástico mal gerenciado por pessoa (kg por ano) na america do sul')
plt.ylabel('')
plt.xlabel('')

plt.show()
```
