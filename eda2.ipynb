```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Analise Exploratoria
Poluição de plásticos no mar

```python
df = pd.read_csv('./datasets/2.csv')
```

```python
df.head(10)
```

```python
df.describe()
# Coluna ano nao utilizada
```

```python
df.info()
```

```python
# Percebi que há uma quantidade por continentes e por pais.
continent = df[(df['Código'].isna()) & (df['Participação na emissão global de plásticos para o oceano'] > 0.1)]
continent = continent.sort_values(by='Participação na emissão global de plásticos para o oceano', ascending=False)
continent
```

```python
plt.figure(figsize=(9, 5))
sns.barplot(x=continent['Entidade'], y=continent['Participação na emissão global de plásticos para o oceano'], palette= 'bright', hue= continent['Entidade'])

plt.title('Participação na emissão global de plásticos para o oceano por continentes no ano de 2019')
plt.ylabel('')
plt.xlabel('')

plt.show()
```

```python
# Top 10 paises que mais jogam plastico no oceano

top10 = df[df['Código'].notnull()].sort_values(by='Participação na emissão global de plásticos para o oceano', ascending=False).head(10)
top10
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=top10['Entidade'], y=top10['Participação na emissão global de plásticos para o oceano'], palette= 'bright', hue= top10['Entidade'])

plt.title('Participação na emissão global de plásticos para o oceano por paises no ano de 2019')
plt.ylabel('')
plt.xlabel('')

plt.show()
```

```python
# Top 5 paises dentro do continente America do Sul

#Encontrando todos os paises e salvando numa variavel

df_sa = df[df['Entidade'].isin(['Brazil', 'Argentina','Paraguay','Uruguay','Peru','Chile','Colombia','Venezuela','Ecuador','French Guiana','Suriname','Trinidad and Tobago', 'Aruba','Guyana'])].sort_values(by='Participação na emissão global de plásticos para o oceano', ascending=False)
df_sa
```

```python
plt.figure(figsize=(18, 5))
sns.barplot(x='Entidade', y='Participação na emissão global de plásticos para o oceano', data=df_sa, hue=df_sa['Entidade'], palette='bright')
plt.title('Barplot das Categorias A e B')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
